# 엑셀 파일 속성값을 개별 리스트에 누적

```python
# 엑셀 파일 1개의 모든 속성값을 개별 리스트에 추가

xml_names = os.listdir(targetdir)

taste_score = []
amount_score = []
delivery_score = []
order_all = []
order_date = []
customer_review_all = []
owner_reivew_all = []
owner_date =[]


for i in tqdm(xml_names):
    csv_name = "../도레미파솔 csv 댓글/" +  i
    df = pd.read_csv(f'{csv_name}', encoding='utf-16')

    taste_score.extend(df['taste'].tolist())
    amount_score.extend(df['amount'].tolist())
    delivery_score.extend(df['delivery'].tolist())
    order_all.extend(df['order'].tolist())
    order_date.extend(df['order_date'].tolist())
    customer_review_all.extend(df['customer'].tolist())
    owner_reivew_all.extend(df['owner'].tolist()) 
    owner_date.extend(df['owner_date'].tolist())
```

# 누적된 리스트의 길이 체크

```python
# 전체 데이터 체크, 컬럼별  갯수 맞는지 체크
print("taste           길이는 " +  str(len(taste_score)))
print("amount          길이는 " +  str(len(amount_score)))
print("delivery        길이는 " +  str(len(delivery_score)))
print("order           길이는 " +  str(len(order_all)))
print("order_date      길이는 " +  str(len(order_date)))
print("customer_review 길이는 " +  str(len(customer_review_all)))
print("owner_reivew    길이는 " +  str(len(owner_reivew_all)))
print("owner_date      길이는 " +  str(len(owner_date)))
```

# 데이터 프레임에 통합 후 생성

```python
#데이터 프레임에 통합해 담기

list_for_df = [taste_score, 
               amount_score, 
               delivery_score, 
               order_all, 
               order_date, 
               customer_review_all,
               owner_reivew_all, 
               owner_date]

# list_for_df = ['taste' : taste_score, 
#                'amount' : amount_score, 
#                'delivery' : delivery_score, 
#                'order' : order_all, 
#                'order_date' : order_date, 
#                'customer_review' :customer_review_all,
#                'owner_reivew' : owner_reivew_all, 
#                'owner_date' : owner_date]

df_all = pd.DataFrame(list_for_df)
df_all = df_all.T
df_all.columns=['taste', 'amount', 'delivery', 
                'order', 'order_date', 'customer_review', 
                'owner_reivew', 'owner_date']
raw_data = df_all
df_all.head()
```

# customer_review 에 결측치 있는 행 제거

```python
df_total = len(df_all)
df_all.dropna(subset=['customer_review']) 
df_all.head()
```

# customer_review 에 0 값이 있으면 행 제거

```python
idx = df_all[df_all['customer_review']=='0'].index
df_all = df_all.drop(idx)
df_all = df_all.reset_index(drop=True)
# df_all.head()
df_all
```

# 맛 긍정 부정

```python
# 1,2 가져오고
# 4,5 가져오고
# 0,1 로 변환

neg_df_1 = df_all[df_all['taste']==1]
neg_df_2 = df_all[df_all['taste']==2]
neg_df_1 = neg_df_1[['taste','customer_review']]
neg_df_2 = neg_df_2[['taste','customer_review']]

pos_df_4 = df_all[df_all['taste']==4]
pos_df_5 = df_all[df_all['taste']==5]
pos_df_4 = pos_df_4[['taste','customer_review']]
pos_df_5 = pos_df_5[['taste','customer_review']]
```

```python
taste_df = pd.concat([neg_df_1,neg_df_2,pos_df_4,pos_df_5])
taste_df
```

```python
# 0 부정, 1긍정 
# 레이블링
taste_df['taste'] = taste_df['taste'].replace(1,0)
taste_df['taste'] = taste_df['taste'].replace(2,0)
taste_df['taste'] = taste_df['taste'].replace(4,1)
taste_df['taste'] = taste_df['taste'].replace(5,1)
```

```python
# 긍정 부정 데이터 셋 크기 확인
neg_sum = taste_df[taste_df['taste']==0]
pos_sum = taste_df[taste_df['taste']==1]
print(len(neg_sum))
print(len(pos_sum))
```

```python
# 부정 갯수에 맞춰서 긍정 갯수 뽑기
pos_sum = pos_sum.sample(n=40000).reset_index(drop=True)
len(pos_sum)
```

```python
# 부정 비율에 맞춘 긍정 댓글 합치기
# 인뎃스 리셋
# 행 썪기
trainset_df = pd.concat([pos_sum,neg_sum]).sample(frac=1).reset_index(drop=True)
len(trainset_df)
```

```python
# 데이터 프레임 확인
trainset_df
```

```python
trainset_df.to_csv('TasteScore_TrainDataSet.csv', encoding='utf-16')
```
